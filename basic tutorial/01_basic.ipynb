### Types in Python

```python
## Type Inference - possible to assing different types to same varialbe
variable = 10
print (type(variable))
variable = "John"
print (type(variable))
```

```python
## type conversion - possible to use int, float and str in some cases for type conversion
variable = 10
print (type(variable))
variable = str(variable)
print (type(variable))
variable = float(variable)
print (type(variable))
variable = int(variable)
print (type(variable))
variable = bool(variable)
print (type(variable))

variable = 0
variable = bool(variable)

print (type(variable))

print(variable)
```

```python
## use for input from user - the defalut type is string, but can be converted
age = input("Enter your age: ")
# age will be string
print("age =", age, ", type =", type(age))

age = int(input("Enter your age: "))
# age will be int
print("age =", age, ", type =", type(age))
```

### Operators

```python
## operators
a = 10
b = 5

print(a + b) # addition
print(a - b) # subtraction
print(a * b) # multiplication
print(a / b) # division
print(a // b) # floor division
print(a % b) # modulo   
print(a ** b) # power
```

### Logical Opertors

```python
## and, or ,not - need to use the words "and", "or" and "not"
a = 10
b = 5

print(a > b) # True
print(a < b) # False
print(a >= b) # True
print(a <= b) # False
print(a == b) # False
print(a != b) # True
print(not a > b) # False
print(a > b and a < b) # False
print(a > b or a < b) # True   
```
