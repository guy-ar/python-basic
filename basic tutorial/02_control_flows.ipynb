### Conditional Statements

```python
## if statement, else if, else
age = 17
if age >= 18:
    print("You are old enough to vote!")
elif age == 17:
    print("You can learn to drive!")

else:
    print("You are not old enough to vote yet.")
```

```python
## nested if statement
## odd or even number
number = int(input("Enter a number: "))
if number >= 0:
    print("Number is positive")
    if number % 2 == 0:
        print("Number is even")
    else:
        print("Number is odd")
else:
    print("Number is negative")
    if number % 2 == 0:
        print("Number is even")
    else:
        print("Number is odd")
```

## Loops

### For Loop

```python
## for loop
for i in range(5): ## not include 5 
    print(i)

for i in range(1,6): ## start in 1 and do not include 6
    print(i)

for i in range(1,10,2): ## start in 1 and do not include 10 and step 2
    print(i)

for i in range(10,1,-1): ## start in 10 and do not include 1 and step -1
    print(i)
```

```python
## string is aray of chars - possible to iterate
for letter in "Hello":
    print(letter)
```

### While Loops

```python
## While loop
i = 0
while i < 5:
    print(i)
    i += 1
```

### Loop Control statement

```python
## break - stop the loop
i = 0
while True: ## endless loop but we have exist using break when reach 5
    print(i)
    i += 1
    if i == 5:
        break   

## use continue to skip iteration
i = 0
while i < 5:
    i += 1
    if i % 2 == 0:
        continue
    print(i)
```

```python
## use for pass to skip iteration
for i in range(5):
    if i % 2 == 0:
        pass ## do nothing
    else:
        print(i)
```

```python
## calculate sum of first n natural numbers
limit = int(input("Enter the limit: "))
sum = 0
for i in range(limit + 1):
   sum = sum + i

print("Sum of first %d natural numbers is %d" %(limit,sum)  )
```

```python
limit = int(input("Enter the limit: "))
sum = 0
numerator = 1
while numerator <= limit:
    sum = sum + numerator
    numerator = numerator + 1

print("Sum of first %d natural numbers is %d" %(limit,sum)  )
```

```python
## all Prime numbers between 1 to 100
for num in range(1,101):
    if num > 1:
        for i in range(2,num):
            if (num % i) == 0:
                break
        else:
            print(num)
    else:
        print(num)
    
```
