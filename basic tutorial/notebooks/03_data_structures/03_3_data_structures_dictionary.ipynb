## Data Strctures

### Dictionaries

key value pairs stores as unordered list, key must be unique and immutable

```python
# define dictionary
empty_dict = {}
print(empty_dict)  # Output: {}
my_dict = {'key1': 'value1', 'key2': 'value2'}
print(my_dict)

# define disctionalty with constructor
my_dict2 = dict(key1='value1', key2='value2')

# accessing dictionary values - first option using the key name
# second option using the get method
print(my_dict['key1'])  # Output: value1
print(my_dict.get('key1'))  # Output: value1

# possible to provide default value if key does not exist
print(my_dict.get('key3', 'default_value'))  # Output: default_value
# without default value it will return None
print(my_dict.get('key3'))  # Output: None
```

### Modify Dictionary elements

```python
student = {
    'name': 'John',
    'age': 20,
    'major': 'Computer Science'
}
print(student)  # Output: {'name': 'John', 'age': 20, 'major': 'Computer Science'}

#modify dictionary elements
student['age'] = 30  # Update age
student['major'] = 'Physics'  # Update major

print(student['name'])  # Output: John
print(student['age'])  # Output: 20
print(student['major'])  # Output: Computer Science

# delete dictionary elements
del student['age']  # Delete age
print(student)  # Output: {'name': 'John', 'major': 'Physics'}
```

### Dictionary methods

```python
keys = student.keys()  # Get keys
print(keys)  # Output: dict_keys(['name', 'major'])

values = student.values()  # Get values
print(values)  # Output: dict_values(['John', 'Physics'])

#  get all items in dictioary using items
items = student.items()  # Get key-value pairs
print(items)  # Output: dict_items([('name', 'John'), ('major', 'Physics')])
```

### Shellow copy

```python
student = {
    'name': 'John',
    'age': 20,
    'major': 'Computer Science'
}
print(student)  # Output: {'name': 'John', 'age': 20, 'major': 'Computer Science'}
# shellow copy of dictionary
new_student = student.copy()  # Create a shallow copy
print(new_student)  # Output: {'name': 'John', 'major': 'Physics'}
# add new key-value pair
student['age'] = 25
print(student)  # Output: {'name': 'John', 'major': 'Physics', 'age': 25}
print(new_student)  # Output: {'name': 'John', 'major': 'Physics'}
# using copy created 2 instances of dictionary - if we just use 
# student = new_student - they will point to the same object
# and any change in one will affect the other

# deep copy of dictionary
# insure that alos internal contents are copied and not just the reference
import copy
deep_copied_student = copy.deepcopy(student)  # Create a deep copy
print(deep_copied_student)  # Output: {'name': 'John', 'major': 'Physics'}
# add new key-value pair
student['age'] = 25
print(student)  # Output: {'name': 'John', 'major': 'Physics', 'age': 25}
print(deep_copied_student)  # Output: {'name': 'John', 'major': 'Physics'}
```

### Iterating over dictionaries

```python
# iterating over dictionaries- over keys
for key in student:
    print(key, student[key])  # Output: name John, major Physics, age 25
# iterating over dictionaries- over values
for value in student.values():
    print(value)  # Output: John, Physics, 25
# iterating over dictionaries- over key-value pairs
for key, value in student.items(): # using unpacking
    # key is the first element and value is the second element
    print(f"{key}: {value}")  # Output: name John, major Physics, age 25key, value)  # Output: name John, major Physics, age 25
```

### Nested Dictionaries

```python
#Nested Dictionaries
stidents = {
    'student1': {
        'name': 'John',
        'age': 20,
        'major': 'Computer Science'
    },
    'student2': {
        'name': 'Alice',
        'age': 22,
        'major': 'Mathematics'
    }
}
print(stidents)

# accessing nested dictionary values
print(stidents['student1']['name'])  # Output: John
print(stidents['student2']['major'])  # Output: Mathematics

# iterating over nested dictionaries
for student_id, student_info in stidents.items():
    print(f"ID: {student_id}")
    for key, value in student_info.items():
        print(f"{key}: {value}")
    print()  # Print a newline for better readability
```

### Dictionary Comprehension

```python
squeare_numbers = {x: x**2 for x in range(1, 6)}
print(squeare_numbers)  # Output: {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}

# Dictionary Comprehension with condition
even_squares = {x: x**2 for x in range(1, 11) if x % 2 == 0}
print(even_squares)  # Output: {2: 4, 4: 16, 6: 36, 8: 64, 10: 100}
```
