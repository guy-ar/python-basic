### Count and presnet the number of instances in list

```python
numbers = [1,4,4,4, 1,1,2,2,2,3,3,3, 5,5,5,5,3, 4, 5]
count_disctionary = {}
for number in numbers:
    if number in count_disctionary: # go over the keys in the dictionary
        count_disctionary[number] += 1
    else:
        count_disctionary[number] = 1
for key in count_disctionary:
    print(str(key) + " paeras in the list: " + str(count_disctionary[key]) + " times")
```

### merge 2 dictionaries

```python
dict1 = {'a': 1, 'b': 2}
dict2 = {'b': 3, 'c': 4}
dict3 = {**dict1, **dict2} ## Merge two dictionaries - as going over the keys in the dictionary
# the key 'b' will be updated to the value 3 from dict2 -it will be the last one
print(dict3)
```

 ### student grades list example

```python
# orgenize students grades
grades = [72, 85, 90, 78, 88, 92, 85, 76]
def calculate_average(grades):
    return sum(grades) / len(grades) # sum and len are built-in functions in Python
print(calculate_average(grades)) 

def find_max_grade(grades):
    return max(grades)  # max is a built-in function in Python
print(find_max_grade(grades))

def find_min_grade(grades):
    return min(grades)  # min is a built-in function in Python
print(find_min_grade(grades))
```
