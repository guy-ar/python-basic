### Tuples

similar to lists but immutable - used for data and constants 
slightly faster then lists

```python
# creating tuple
empty_tuple = ()
print(empty_tuple)  # Output: ()
print(type(empty_tuple))  # Output: <class 'tuple'>
single_element_tuple = (1,)  # Note the comma
```

```python
# another way to create a list or tuple
empry_list = list()  # Using list constructor
print(empry_list)  # Output: []
print(type(empry_list))  # Output: <class 'list'>
empty_tuple = tuple()  # Using tuple constructor
print(empty_tuple)  # Output: ()
print(type(empty_tuple))  # Output: <class 'tuple'>
single_element_tuple = tuple([1])  # Using tuple constructor
print(single_element_tuple)  # Output: (1,)
print(type(single_element_tuple))  # Output: <class 'tuple'>
single_element_list = list([1])  # Using list constructor
print(single_element_list)  # Output: [1]    
print(type(single_element_list))  # Output: <class 'list'>
```

```python
# Convert list to tuple
tuple_from_list = tuple(single_element_list)
print(tuple_from_list)  # Output: (1,)

# Convert tuple to list
list_from_tuple = list(single_element_tuple)
print(list_from_tuple)  # Output: [1]

# creating tuple from a list
list_of_numbers = [1, 2, 3, 4, 5]
tuple_of_numbers = tuple([1, 2, 3, 4, 5]) # Convert list to tuple
print(tuple_of_numbers)  # Output: (1, 2, 3, 4, 5)

# creatting list from a tuple
tuple_of_numbers = (1, 2, 3, 4, 5)
list_of_numbers = list((1, 2, 3, 4, 5)) # Convert tuple to list
print(list_of_numbers)  # Output: [1, 2, 3, 4, 5]
```

### Tuples Indexing and slicing - like lists

```python
# Accessing elements in a tuple
tuple_of_numbers = (1, 2, 3, 4, 5)
print(tuple_of_numbers[0])  # Output: 1
print(tuple_of_numbers[1])  # Output: 2
print(tuple_of_numbers[-1])  # Output: 5 - last element
print(tuple_of_numbers[0:3])  # Output: (1, 2, 3) - first three elements
print(tuple_of_numbers[::-1])  # Output: (5, 4, 3, 2, 1) - reversed tuple
```

### Tuple Operations

```python
# as thery are immutalbe they do not support item assignment
# tuple_of_numbers[0] = 10  # This will raise a TypeError: 'tuple' object does not support item assignment

# tuple concatenation
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
concatenated_tuple = tuple1 + tuple2
print(concatenated_tuple)  # Output: (1, 2, 3, 4, 5, 6)

# can multiply tuples
tuple3 = (1, 2, 3)
multiplied_tuple = tuple3 * 3
print(multiplied_tuple)  # Output: (1, 2, 3, 1, 2, 3, 1, 2, 3) 
# counting elements in a tuple
tuple_of_numbers = (1, 2, 3, 1, 2, 1)
print(tuple_of_numbers.count(1))  # Output: 3 - counts how many times 1 appears
print(tuple_of_numbers.count(2))  # Output: 2 - counts how many times 2 appears

# tuple length
tuple_of_numbers = (1, 2, 3, 4, 5)
print(len(tuple_of_numbers))  # Output: 5

# tupple index
tuple_of_numbers = (1, 2, 3, 4, 5)
print(tuple_of_numbers.index(3))  # Output: 2 - index of the first occurrence of 3
# print(tuple_of_numbers.index(6))  # Raises ValueError: tuple.index(x): x not in tuple
# tuple membership
tuple_of_numbers = (1, 2, 3, 4, 5)  
print(3 in tuple_of_numbers)  # Output: True
print(6 in tuple_of_numbers)  # Output: False

# tuple packing
packed_tuple = 1, 2, 3  # Packing values into a tuple - not using parentheses
print(packed_tuple)  # Output: (1, 2, 3)

# tuple unpacking
tuple_of_numbers = (1, 2, 3)
a, b, c = tuple_of_numbers ## Unpacking values from a tuple
print(a)  # Output: 1
print(b)  # Output: 2
print(c)  # Output: 3

# unpacking with * operator
tuple_of_numbers = (1, 2, 3, 4, 5)
a, *b, c = tuple_of_numbers ## Unpacking values from a tuple
print(a)  # Output: 1
print(b)  # Output: [2, 3, 4]
print(c)  # Output: 5
```

### Nested Tuples

```python
# like nested lists, tuples can also be nested
nested_tuple = (1, (2, 3), (4, 5))
print(nested_tuple)  # Output: (1, (2, 3), (4, 5))

# accessing nested tuples
print(nested_tuple[1])  # Output: (2, 3) - accessing

# iterating through a nested tuple
for item in nested_tuple:
    if isinstance(item, tuple):
        for sub_item in item:
            print(sub_item, end=' ')  # Output: 2 3 4 5
    else:
        print(item, end=' ')  # Output: 1
```
