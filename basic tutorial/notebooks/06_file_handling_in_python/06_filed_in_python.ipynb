## Read and write files in python

```python
# Open file in read mode and get his content
with open('example.txt', 'r') as file:
    content = file.read()    
    print(content)
```

```python
# open file in readd mode and get lines in file
with open('example.txt', 'r') as file:
    for line in file:
        print(line.strip())  # Print each line without extra newline characters
```

```python
# another option to read lines
with open('example.txt', 'r') as file:
    lines = file.readlines()  # Read all lines into a list
    for line in lines:
        print(line.strip())  # Print each line without extra newline characters
```

```python
# open flie in write mode and write to it - it will overwrite the file
with open('example.txt', 'w') as file:
    file.write("This is a new line added to the file.\n")
    file.write("Another line added to the file.\n")
```

```python
# open file for write in append mode and add new lines to it
with open('example.txt', 'a') as file:
    file.write("This is an additional line added to the file.\n")
    file.write("Yet another line added to the file.\n")
```

```python
# write multuple lines to file in append mode
with open('example.txt', 'a') as file:
    lines_to_add = [
        "This is a new line added to the file.\n",
        "Another line added to the file.\n",
        "This is an additional line added to the file.\n",
        "Yet another line added to the file.\n"
    ]
    file.writelines(lines_to_add)  # Write multiple lines at once
```

### Work with Binary files

```python
# in binary files we are update in byte mode - read byte or write byte
# open binary file in read mode and get his content
with open('example.txt', 'rb') as file:
    content = file.read()
    print(content)
    
# open binary file in write mode and write to it - it will overwrite the file
with open('example.txt', 'wb') as file:
    file.write(b"This is a new line added to the file.\n")
```

```python
# read from source file and write to destination file
with open('source.txt', 'rb') as source_file:
    with open('destination.txt', 'wb') as destination_file:
        content = source_file.read()
        destination_file.write(content)  # Write the content to the destination file        
## and in append mode
# read from source file and write to destination file
with open('source.txt', 'rb') as source_file:
    with open('destination.txt', 'ab') as destination_file:
        content = source_file.read()
        destination_file.write(content)  # Write the content to the destination file
```

```python
### using w+ mode
# open file in read and write mode and get his content
# if the file does not exist, it will be created
with open('not_exist.txt', 'w+') as file:
    content = file.read()  # This will not work if the file is empty
    print(content)
    file.write("This is a new line added to the file.\n")
    file.seek(0)  # Move the cursor back to the beginning of the file
    content = file.read()  # Now we can read the content
    print(content)
```
