## Standard Libraries in Python

```python
## use of array in Python
import array
# Create an array of integers
int_array = array.array('i', [1, 2, 3, 4, 5]) # 'i' indicates the typecode for signed integers
# Print the array
print("Integer Array:", int_array)
```

```python
## use of random in Python
import random
# Generate a random integer between 1 and 10
random_integer = random.randint(1, 10)
# Generate a random floating-point number between 0 and 1
random_float = random.random()
# Print the generated values
print("Random Integer:", random_integer)
print("Random Float:", random_float)

random.choice([1, 2, 3, 4, 5])  # Randomly choose an element from the list 
```

### File and Directory access

```python
##  use os in Python
import os
# Get the current working directory
current_directory = os.getcwd()
# Print the current directory
print("Current Directory:", current_directory)
# List all files and directories in the current directory
files_and_directories = os.listdir(current_directory)
# Print the list of files and directories
print("Files and Directories:", files_and_directories)
## make a new directory
new_directory = "new_folder"
os.mkdir(new_directory)
```

### Data Seriliazition

```python
# use of jason in Python
import json
# Create a sample dictionary
data = {    "name": "Alice",
            "age": 30, 
            "city": "New York"
        }
# Convert the dictionary to a JSON string
json_string = json.dumps(data)
# Print the JSON string
print("JSON String:", json_string)
print(type(json_string))

# Convert the JSON string back to a dictionary
data_from_json = json.loads(json_string)
# Print the dictionary
print("Dictionary from JSON:", data_from_json)
print(type(data_from_json))
```

### Use for csv

```python
# use csv in Python
import csv
# Create a sample CSV file
csv_data = [    ["Name", "Age", "City"],
                ["Alice", 30, "New York"],  
                ["Bob", 25, "Los Angeles"],
                ["Charlie", 35, "Chicago"] 
            ]
# Write the data to a CSV file 
with open('sample.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(csv_data)
# Read the data from a CSV file
with open('sample.csv', mode='r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

### use of Datatime

```python
# use for datetime in Python
from datetime import datetime, timedelta
# Get the current date and time
current_datetime = datetime.now()
# Print the current date and time   
print("Current Date and Time:", current_datetime)
# Add 7 days to the current date
future_date = current_datetime + timedelta(days=7)
# Print the future date
print("Future Date (7 days later):", future_date)
# Format the current date as a string
formatted_date = current_datetime.strftime("%Y-%m-%d %H:%M:%S")
# Print the formatted date
print("Formatted Date:", formatted_date)

# use of time in Python
import time
# Get the current time in seconds since the epoch
current_time = time.time()
# Print the current time
print("Current Time (in seconds):", current_time)
# Get the current time in milliseconds
current_time_ms = time.time() * 1000
# Print the current time in milliseconds
print("Current Time (in milliseconds):", current_time_ms)

time.sleep(2)  # Pause the program for 2 seconds
print("Program paused for 2 seconds.")
current_time = time.time()
# Print the current time
print("Current Time (in seconds):", current_time)
```
