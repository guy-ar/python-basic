## Import  modules and packages in Python
## Common packages to use

```python
## In order to use existing packages need to import them
# Python has a rich ecosystem of libraries and modules that can be imported to extend functionality.
# Importing modules allows you to use pre-written code, functions, and classes without having to write them from scratch.

import math  # Importing the math module for mathematical functions
print("Square root of 16 is:", math.sqrt(16))  # Output: Square root of 16 is: 4.0

import random  # Importing the random module for generating random numbers
from random import randint  # Importing a specific function from the random module
print("Random number between 1 and 10:", randint(1, 10)) ## no need to use random.randint

import numpy as np  # Importing the NumPy library for numerical operations
print("Numpy version:", np.__version__)  # Output: Numpy version:
np.array([1, 2, 3])  # Creating a NumPy array

## Importing specific functions from a module
from math import pi, sin  # Importing specific functions from the math module
print("Value of pi:", pi)  # Output: Value of pi: 3.141592653589793

# importing all functions from a module (not recommended)
from math import *  # Importing all functions from the math module (not recommended)
print("Cosine of 0 is:", cos(0))  # Output: Cosine of 0 is: 1.0
```

## Create your own package

```python
## start with a module by createing a directory with the same name as the module
# Create a directory named 'mymodule' and add an __init__.py file inside it
# then create a file named mymodule.py inside the 'mymodule' directory
# The __init__.py file can be empty or contain initialization code for the package.
# then we can crate a file named mymodule.py inside the 'mymodule' directory
# The mymodule.py file can contain functions, classes, and variables that you want to include in the package
# Example content of mymodule.py
# def greet(name):
#     print(f"Hello, {name}!")  
from mymodule.mypackage import greet  # Importing the greet function from mymodule
greet("Alice")  # Output: Hello, Alice!

# posible also to create sub-packages by creating additional directories inside the package directory
# For example, if you have a directory structure like this:
# mymodule
# └── mypackage
#     ├── __init__.py
#     └── subpackage
#         └── __init__.py
# you can import the subpackage like this:
# from mymodule.mypackage.subpackage import some_function
```
